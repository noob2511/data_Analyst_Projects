# Maksed Array
The masked array is the arrays that have invalid or missing entries. It can be created by using the masked_array function. It's use is to create a new array full of masked values.

```python
a = np.array([1,2,3,4,-1])
b = ma.masked_array(a , mask = [0,0,0,0,-1])
print(b)
```

# Types of arrays:
1. 1 - D array : It is essentially a list of elements. It can contain either a single row or a single column. The shape of 1-D array is in `n,` where n is the number of elements. 
2. 2- D array : It is like a matrix or a table with rows and columns. It has a shape of `m,n` where `m` is the number of rows and `n` is the number of columns
3. 3-D array : 3-D array is a collection of 2-D array. It has a shape of `p,m,n`, where `p` is the number of 2-D arrays, `m` is the row number and `n` is the column number

```python
#Creation of 1 - D array
array_1D = np.array([1,2,3,4,5])
print('1 D array : ', array_1D)
print(array_1D.shape)
print()

#Creation of 2 - D array

array_2D = np.array([[1,2],[3,4], [5,6]])
print('2 D array : ', array_2D)
print(array_2D.shape)
print()

#Creation of 3 - D array

array_3D = np.array([[[1,2],[3,4], [5,6],[7,8]]])
print('3 D array : ', array_3D)
print(array_3D.shape)
print()
```

# Accessing values of arrays
Accessing values in arrays is done in the same way as we do for list by providing indexes in the form of `array[n]`, where n represents the value of accesible index.

```python
#Accessing values for 1-D array
element_1D = array_1D[2]
print('The element for 1-D array : ', element_1D)
print()

#Accessing values for 2-D array
element_2D = array_2D[2,0]
print('The element for 2-D array : ', element_2D)
print()

#Accessing values for 3-D array
element_3D = array_3D[0,0,0]
print('The element for 3-D array : ', element_3D)
print()
```

# Random function
Random function in numpy is used to create arrays filled with random values. There are 3 ways by which we can create a random array:
1. **By using rand function :** We can create a random array between `0 and 1` and it will consider floating point values and the range will be exclusive of `0 and 1`.
2. **By using randn function :** We can create a random array of any value normally and it can take positive and negative values as well. It will consider floating point values only.
3. **By using uniform function :** We can create a random uniform array by giving it a size as well as a range


```python
#Example of rand function

array_1D = np.random.rand(5)
print('1-D array : ',array_1D)
print()

array_2D = np.random.rand(2,2)
print('2-D array : ',array_2D)
print()
array_3D = np.random.rand(2,3,4)
print('3-D array : ',array_3D)
print()
```

```python
#Example of randn function

array_1D = np.random.randn(5)
print('1-D array : ',array_1D)
print()

array_2D = np.random.randn(2,2)
print('2-D array : ',array_2D)
print()
array_3D = np.random.randn(2,3,4)
print('3-D array : ',array_3D)
print()
```

```python
#Example of random uniform function

array_1D = np.random.uniform(5, size = 2)
print('1-D array : ',array_1D)
print()

array_2D = np.random.uniform(2,2, size = (2,2))
print('2-D array : ',array_2D)
print()
array_3D = np.random.uniform(2,3, size =(2,3,4))
print('3-D array : ',array_3D)
print()
```

# Randint function
Randint function is used to print random `int` values

```python
#Example of randint function

array_1D = np.random.randint(5, size = 2)
print('1-D array : ',array_1D)
print()

array_2D = np.random.randint(0,10, size = (2,2))
print('2-D array : ',array_2D)
print()
array_3D = np.random.randint(0,10, size =(2,3,4))
print('3-D array : ',array_3D)
print()
```
